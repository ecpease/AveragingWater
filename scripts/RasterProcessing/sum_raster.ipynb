```python
import os
import numpy as np
import os
import rasterio
import pandas as pd

directory = os.path.join("..", "output_rasters")

years = []
vols  = []

for filename in os.listdir(directory):
    if filename.endswith(".tif"): 
        raster = rasterio.open(os.path.join(directory, filename))
        arr = raster.read(1)
        add = np.nansum(arr)
        years.append(filename[-10:-4]) # Grabs the year from the filename
        vols.append(add)
        d = {'Year': years, 'AcreFt': vols} # Create dataframe from the array
        df = pd.DataFrame(data=d)
        df.sort_values(by=['Year'], inplace=True) # Sort years small to large
```

```python
df.to_csv(os.path.join("..", "Data", "RiverSums.csv"))
```
